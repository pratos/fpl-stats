### Overall players leading by xA, xG

#### What is xG and xA? (Link: [Fbref.com](https://fbref.com/en/expected-goals-model-explained/))
> What is xG?
Very simply, xG (or expected goals) is the probability that a shot will result in a goal based on the characteristics of that shot and the events leading up to it. Some of these characteristics/variables include: 

* Location of shooter: How far was it from the goal and at what angle on the pitch?
* Body part: Was it a header or off the shooter's foot?
* Type of pass: Was it from a through ball, cross, set piece, etc?
* Type of attack: Was it from an established possession? Was it off a rebound? Did the defense have time to get in position? Did it follow a dribble?

```python
fpl_df.groupby(by=['team_slug', 'display_slug', 'position_fbref'], group_keys=False, sort=True)\
    .agg({'xg': 'sum', 'goals_scored': 'sum', 'xa': 'sum', 'assists_fpl': 'sum', 
          'position_fpl': 'sum', 'total_points': 'sum', 'cost': 'sum'})\
    .sort_values(by=['total_points', 'xg', 'xa'], ascending=False)\
    .reset_index()\
    .head(20)
```

### Performers by position

#### Forwards

```python
fpl_df.groupby(by=['team_slug', 'display_slug', 'position_fbref'], group_keys=False, sort=True)\
    .agg({'xg': 'sum', 'goals_scored': 'sum', 'xa': 'sum', 'assists_fpl': 'sum', 
          'position_fpl': 'sum', 'total_points': 'sum', 'cost': 'sum'})\
    .sort_values(by=['total_points', 'xg', 'xa'], ascending=False)\
    .reset_index()\
    .query('position_fpl in ["FWD", "FWDFWD"]')\
    .head(20)
```

#### Midfielders

```python
fpl_df.groupby(by=['team_slug', 'display_slug', 'position_fbref'], group_keys=False, sort=True)\
    .agg({'xg': 'sum', 'goals_scored': 'sum', 'xa': 'sum', 'assists_fpl': 'sum', 
          'position_fpl': 'sum', 'total_points': 'sum', 'cost': 'sum'})\
    .sort_values(by=['total_points', 'xg', 'xa'], ascending=False)\
    .reset_index()\
    .query('position_fpl in ["MID", "MIDMID"]')\
    .head(20)
```

#### Defenders (A few more params here will make sense)

```python
fpl_df.groupby(by=['team_slug', 'display_slug', 'position_fbref'], group_keys=False, sort=True)\
    .agg({'xg': 'sum', 'goals_scored': 'sum', 'xa': 'sum', 'assists_fpl': 'sum', 
          'position_fpl': 'sum', 'total_points': 'sum', 'cost': 'sum'})\
    .sort_values(by=['total_points', 'xg', 'xa'], ascending=False)\
    .reset_index()\
    .query('position_fpl in ["DEF", "DEFDEF"]')\
    .head(20)
```

#### Goalkeepers

#### Univariate

```python
fpl_grouped_df = fpl_df.groupby(by=['team_slug', 'position_fbref'], group_keys=False, sort=True)\
    .agg({'xg': ['sum', 'mean', 'max', 'min'], 'xa': ['sum', 'mean', 'max', 'min']})
fpl_grouped_df.columns = ['xg', 'xg_mean', 'xg_max', 'xg_min', 'xa', 'xa_mean', 'xa_max', 'xa_min']
    
fpl_grouped_df.sort_values(by=['xg', 'xa'], ascending=False)\
    .reset_index()\
    .head(20)
```
